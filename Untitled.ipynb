### Step 1: Load the Data

```python
import pandas as pd

# Load dataset
df = pd.read_csv("Mall_Customers.csv")
print(df.head())
```

### Step 2: Understand the Data (Exploratory Data Analysis)

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gender distribution
sns.countplot(x='Gender', data=df)
plt.show()

# Age vs Spending
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)', hue='Gender')
plt.show()
```

### Step 3: Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

# Encode 'Gender'
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])  # Male:1, Female:0

# Optional: drop CustomerID if not useful
X = df.drop(['CustomerID'], axis=1)  # use all other features
```

### Step 4: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Step 5: Determine Optimal Number of Clusters (Elbow Method)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.grid()
plt.show()
```

### Step 6: Apply K-Means Clustering with k = 5

```python
from sklearn.cluster import KMeans

# Fit KMeans with 5 clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Add cluster labels to original dataframe
df['Cluster'] = y_kmeans
```

### Step 7: Visualize the Clusters

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', 
                hue='Cluster', palette='Set1', s=100)
plt.title('Customer Segmentation')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

# Reduce to 2D for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=df['Cluster'], palette='Set2', s=100)
plt.title('Customer Segmentation (PCA Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.grid(True)
plt.show()
```

### Step 8: Analyze & Interpret the Segments

```python
cluster_summary = df.groupby('Cluster').mean(numeric_only=True)
print(cluster_summary)
```

```python
for cluster_num in range(5):
    print(f"\nCluster {cluster_num}:")
    print(df[df['Cluster'] == cluster_num].describe())
```

###  Step 9: Evaluate Clustering Quality (Silhouette Score)

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, y_kmeans)
print(f"Silhouette Score: {score:.3f}")
```
